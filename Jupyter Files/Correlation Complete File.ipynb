```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from collections import OrderedDict 
import math
import numpy as np
import io 
```

```python
data1=pd.read_csv('2 EG_FEC_RNEW.csv',index_col='S.No.')
data1=data1.drop(columns=['2001'])
data1=data1.rename(columns={'2017':'Renewable energy share in the total final energy consumption (%)'})
data1
```

```python
data2=pd.read_csv('4 EG_EGY_RNEW.csv',index_col='S.No.')
data2=data2.drop(columns=['2001'])
data2=data2.rename(columns={'2017':'Installed renewable electricity-generating capacity (watts per capita)'})
data2
```

```python
common=pd.merge(data1, data2, how ='inner', on =['GeoAreaName'])
common
```

```python
correlation=(common['Renewable energy share in the total final energy consumption (%)'].corr(common['Installed renewable\xa0electricity-generating capacity (watts per capita)']))*100
if 20<=correlation<=80:
  print("Moderate Correlation")
elif correlation>80:
  print("Strong Correlation")
else:
  print("Weak Correlation")
print("Correlation between these data is:",round(correlation,2),"%")
```
